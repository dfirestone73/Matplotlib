```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cityData=pd.read_csv(city_data_to_load)
rideData=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
totalData=pd.merge(cityData,rideData,on="city",how="outer")

# Display the data table for preview
totalData.head()
```

## Bubble Plot of Ride Sharing Data

```python
dictList=[]


for city in totalData["city"].unique():
    dictionary={"City":totalData.loc[totalData["city"]==city]["city"].max(),
                "Driver Count":totalData.loc[totalData["city"]==city]["driver_count"].max(),
                "Type":totalData.loc[totalData["city"]==city]["type"].max(),
                "Total Fares":totalData.loc[totalData["city"]==city]["fare"].sum(),
                "Avg. Fare":totalData.loc[totalData["city"]==city]["fare"].mean(),
                "Total Rides":totalData.loc[totalData["city"]==city]["type"].count()
                }
    dictList.append(dictionary)

summaryData=pd.DataFrame(dictList)
summaryData=summaryData.set_index("City")
maxRides=summaryData["Total Rides"].max()
maxFare=summaryData["Avg. Fare"].max()
```

```python
# Obtain the x and y coordinates for each of the three city types
x_axis=summaryData["Total Rides"]
y_axis=summaryData["Avg. Fare"]
plt.xlabel("Total Number of Rides (Per city)")
plt.ylabel("Average Fare ($)")
sizes = summaryData["Driver Count"]
colors=[]
for color in summaryData["Type"]:
    if color=="Urban":
        colors.append("coral")
    elif color=="Suburban":
        colors.append("skyblue")
    elif color=="Rural":
        colors.append("gold")
# Build the scatter plots for each city types

plt.scatter(x_axis,
             y_axis,
             s=sizes*10,
             color=colors,
             edgecolors="black"
             )
# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data (2016)')
plt.grid()

# Create a legend
labels=[['Urban','coral'],['Suburban','skyblue'],['Rural','gold']]
for area in labels:
    plt.scatter([],[],c=area[1],alpha=1,s=10,label=area[0])
    
plt.legend(scatterpoints=1,frameon=True,labelspacing=.5,title='Type')
plt.legend(loc='upper right')

# Incorporate a text label regarding circle size
plt.text(42,35, 'Note: Circle size correlateswith driver count per city.', horizontalalignment='left',
      verticalalignment='center',wrap=True)

# Save Figure
plt.savefig("pyberScatterPlot.png")
```

## Total Fares by City Type

```python
pieCharts=summaryData.set_index('Type')
pieCharts=pieCharts.groupby('Type')['Total Fares','Total Rides','Driver Count'].sum()
```

```python
# Calculate Type Percents


# Build Pie Chart
labels=['Rural','Suburban','Urban']
colors=['gold','skyblue','coral']
sizes=pieCharts["Total Fares"]
explode=(0,0,.1)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",startangle=100)
plt.title('% of Total Fares by City Type')
plt.show()
# Save Figure
plt.savefig("pyberFares.png")
```

```python
# Calculate Ride Percents

# Build Pie Chart
labels=['Rural','Suburban','Urban']
colors=['gold','skyblue','coral']
sizes=pieCharts["Total Rides"]
explode=(0,0,.1)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",startangle=100)
plt.title('% of Total Rides by City Type')
plt.show()
# Save Figure
plt.savefig("pyberRides.png")
```
